```python
import pandas as pd
```

```python
df1 = pd.read_csv("Lib_2014_dates_only.csv")
```

```python
df2 = pd.read_csv("Lib_2015_dates_only.csv")
```

```python
df3 = pd.read_csv("Lib_2016_dates_only.csv")
```

```python
df3
```

```python
concatenated_df = pd.concat([df1, df2, df3], axis=0,)
```

```python
concatenated_df['DateOnly'] = pd.to_datetime(concatenated_df['DateOnly'])
```

```python
concatenated_df.groupby(concatenated_df['DateOnly'].dt.day)
```

```python
Same_dates = concatenated_df.groupby([concatenated_df['DateOnly'].dt.day, concatenated_df['DateOnly'].dt.month_name()]).sum()
```

```python
Same_dates
```

```python
Same_dates.to_csv('comb_dates.csv')
```
