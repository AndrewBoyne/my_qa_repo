```python
X_train, X_test, y_train, y_test = train_test_split(df1[Cols],
                                                    df1['Hour'],
                                                    test_size=0.30,
                                                    random_state =42)
```

```python
import numpy as np
```

```python
from itertools import combinations
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
y_true = y_test

results = {}
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
```

```python
r2 = r2_score(y_true, y_pred)
```

```python
results[str(Cols)] = {'RMSE':rmse,
                        'r^2': r2}
```

```python
df1_results = pd.DataFrame.from_dict(results, orient = 'index')
df1_results.reset_index(inplace=True)
df1_results.columns = ['Features','RMSE','R2']
df1_results.sort_values(by='RMSE', inplace=True)
```
