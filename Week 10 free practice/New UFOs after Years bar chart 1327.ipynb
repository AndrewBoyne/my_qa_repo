```python
import pandas as pd
```

```python
#As I was unable to amend the header columnswith the dots so I amended the spread sheet and saved as UFOs
df=pd.read_csv("UFOs.csv") 
```

```python
df.head()
```

```python
df=df.rename(columns={'Encounter duration':'Duration','.Day':'Day'})
```

```python
df.head()
```

```python
# Drop unnecessary columns
df.drop(['Dates.Documented.Year','Dates.Documented.Month','Dates.Documented.Day'], axis=1, inplace=True)
```

```python
df.head()
```

```python
# What are the entries in Shape?
df.groupby(['Shape']).count()
```

```python
# Create new column with totals of the different Shapes
df['Shape_count']=df['Shape'].groupby(df['Shape']).transform('count')
```

```python
df.head()
```

```python
new_columns=['Shape', 'Shape_count']
```

```python
shapes=df[new_columns]
```

```python
shapes.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ax = sns.barplot(x='Shape', y='Shape_count', data = shapes)
plt.xticks(rotation=90)
for p in ax.patches:
    ax.text(p.get_x() + p.get_width() / 2., p.get_height(), str(int(p.get_height())),
            ha='center', va='bottom', visible=False)
```
