```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
#I've imported the tools to be able to do some things - eg; creating a linear regression and r2
```

```python
np.random.seed(0)
x=np.random.rand(50)*10
y=2*x+1+np.random.randn(50)*2
```

```python
#I've created some random data (50 numbers between 0 and 10) and some other random data 
# (subject to a formula)
```

```python
corr_coeff, p_value = pearsonr(x, y)
correlation_text = f"r = {corr_coeff:.2f} and p-value= {p_value}"
# Absolutely no feckin clue
```

```python
x=x.reshape(-1, 1)
y=y.reshape(-1, 1)
```

```python
# changes the x and y axes and, therefore, the size / shape of the graph
```

```python
model= LinearRegression()
model.fit(x, y)
equation_text=f"y={model.coef_[0][0]:.2f}x+{model.intercept_[0]:.2f}"
x_pred=np.linspace(0,10,100).reshape(-1,1)
y_pred=model.predict(x_pred)
```

```python
r_squared=r2_score(y, model.predict(x))
r_squared_text=f"R Squared={r_squared:.2f}"
```

```python
residuals=y-model.predict(x)
std_dev=np.std(residuals)
two_std=2*std_dev
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x, y, label="Data points", color="blue")
plt.plot(x_pred, y_pred, color="red", label="Linear Regression")
plt.fill_between(
    x_pred.flatten(),
    y_pred.flatten() - two_std,
    y_pred.flatten() + two_std,
    color="gray",
    alpha=0.3,
    label="Error Window",
)
plt.legend()
plt.grid(True)
plt.show()
```
