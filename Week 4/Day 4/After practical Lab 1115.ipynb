# Matplotlib (plt) and Seaborn (sns)

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
df1=pd.read_csv('mortgage_applicants.csv')
df1
```

```python
fig, ax=plt.subplots(figsize=(10, 2)) # changes the dimensions of the frame
plt.scatter(df1['Debt'],              # x axis by default
           df1['Balance']);           # y axis by default
```

```python
fig, ax=plt.subplots(figsize=(15, 5)) # changes the dimensions of the frame
plt.scatter(df1['Debt'],
           df1['Balance']);
```

```python
fig, (ax1, ax2)=plt.subplots(1, 2, figsize=(10, 2)) # 2 rows, 2 columns
```
